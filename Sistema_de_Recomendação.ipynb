# **Sistema de Recomendação**


Um Sistema de Recomendação é uma aplicação de software projetada para sugerir itens ou informações relevantes a usuários, com base em suas preferências e comportamentos anteriores. Esses sistemas são amplamente utilizados em várias plataformas online, como sites de comércio eletrônico, serviços de streaming de música e vídeo, redes sociais, entre outros.



# *Tipos de Sistemas de Recomendação*

**Filtragem Colaborativa:** Essa abordagem faz recomendações com base nas preferências de usuários semelhantes. Se um usuário A gosta dos mesmos itens que um usuário B, o sistema pode sugerir a A itens que B tenha gostado e que A ainda não tenha visto.

**Filtragem Baseada em Conteúdo:** Este método analisa as características dos itens e as preferências do usuário para fazer recomendações. Se um usuário demonstrou interesse em certo tipo de conteúdo, o sistema recomendará itens semelhantes em termos de características ou conteúdo.

**Sistemas Híbridos:** Esses sistemas combinam abordagens de filtragem colaborativa e baseada em conteúdo para fornecer recomendações mais precisas e robustas. Isso ajuda a superar algumas limitações inerentes a cada método isoladamente.

**Aprendizado Profundo (Deep Learning):** Algumas implementações avançadas utilizam redes neurais profundas para aprender padrões complexos nos dados de preferência do usuário e características do item, permitindo recomendações mais personalizadas.

# **Quantidade de Avaliação por usuario**

```python
ratings['user-id'].value_counts()
```

```python
# Livros que tenham mais de 200 avaliações
avaliacoes_filtradas = ratings['user-id'].value_counts() > 200
```

```python
#quantidade usuário e user-ids
y=avaliacoes_filtradas[avaliacoes_filtradas].index
print(y.shape)
```

# **Decisão de negócio**


```python
#Trazendo ratings somente do usuario que avaliaram mais de 200 livros
ratings=ratings[ratings['user-id'].isin(y)]
```

```python
#Juntando tabelas (join ou merge)
rating_with_books = ratings.merge(books, on='ISBN')
rating_with_books.head()
```

```python
#quantidade de avaliações dos livros (atributo para contar quantidade vezes de avaliações)
numero_avaliacao = rating_with_books.groupby('title')['rating'].count().reset_index()
```

```python
numero_avaliacao
```

```python
#renomeando coluna Rating
numero_avaliacao.rename(columns= {'rating':'number_of_rating'},inplace=True)
numero_avaliacao
```

```python
#juntando tabela de livros com tabela de quantidade
final_rating = rating_with_books.merge(numero_avaliacao, on ='title')
final_rating
```

# **Decisão de negócio**

```python
#filtragem somente livros pelo menos 50 avaliações
final_rating = final_rating[final_rating['number_of_rating'] >=50]
final_rating.shape
```

```python
#descartando valores duplicados
final_rating.drop_duplicates(['user-id','title'], inplace=True)
final_rating.shape
```

```python
final_rating
```

# **Preparando base dados para modelagem**

```python
#criando pivot usuarios em colunas e avaliações serão máquina preditiva

book_pivot = final_rating.pivot_table(columns='user-id', index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
#subistituir numeros nulos (NAN)
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

Preparando conjunto de dados para modelagem usando o algoritmo de vizinhos mais próximos que é o agrupamento com base na distância Euclidiana



**Biblioteca scipy** é utilizado para lidar eficientemente com matrizes esparsas. Matrizes esparsas são matrizes que contêm uma grande quantidade de elementos zero.

**Tabela Dinâmica** converte para matriz esparsa pois a nossa tabela contém muitos numeros zeros. Para depois alimentar o modelo.

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

# **Máquina preditiva**

https://scikit-learn.org/stable/modules/neighbors.html



```python
from skearn.neghboars import NearestNeighboars
model=NearestNeighboars(algorithm='brute')
model.fit(book_sparse)

```
